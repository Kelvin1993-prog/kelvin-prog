# visualizing the gender gap in college degrees

visualizing the gender gap in college degrees. we only have the women column to work with. we will start by importing the matplotlib inline (%), which informs jupyter that we are working with matplotlib visualization tools.

next step is to import the pandas dataframe, since we are working with data that are represented as csv file which consist of rows and columns(dataframe).
we will then import the matplotlib as plt to import the library for plotting. 


```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv') #reading our csv file with pandas
cb_dark_blue = (0/255,107/255,164/255) #following the RGB color values which should not exceed between 0 and 1
cb_orange = (255/255, 128/255, 14/255) #assingning our orange color line. we now have two color lines for our line plot, which are orange and blue
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics'] # assinging the stem subject category to a variable or seperate folder.

fig = plt.figure(figsize=(18, 3)) # simply, we are telling our jupyter to display a plot figure with width of 18 and height of 3
# it is time for us to display our plot figure for the stem categories,
# but we want to avoid the risk of ambiguity i.e reading each subject seperately and plotting it.
# we will use the for loop to iterate over the stem_cats variable and automatically plot for all subjects.

for sp in range(0,6): # for subjects that falls within range o,6 [index value asssignment] with the stem_cat
    ax = fig.add_subplot(1,6,sp+1) # plot the figure using the range values which increases or iterates automaticallly until the range catrgory is covered
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False) # we are removing the spines to improve the aesthetics of our plot visualization    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011) #setting our x cordinate, variables to be within a certain year period
    ax.set_ylim(0,100)# setting our y cordinate to be in this variable
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```
